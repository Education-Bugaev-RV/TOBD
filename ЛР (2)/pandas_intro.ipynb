```python
review_points_mean = wine_review.points.mean()
review_points_mean
```

```python
wine_review.points.map(lambda p: p - review_points_mean)
wine_review.points
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

wine_mean = wine_review.apply(remean_points, axis='columns')
```
