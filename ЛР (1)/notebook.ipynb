## CSV:

### Задание


Подсчитать число выздоровевших в каждой стране за февраль. Вывод в формате «страна-число»

```python
df_raw_csv = pd.read_csv('./Вариант 4/covid_19_data.csv')

# приводит данные в колонке date к datetime
df_raw_csv['ObservationDate'] = pd.to_datetime(df_raw_csv['ObservationDate'])

# отбирает данные за февраль
df_february = df_raw_csv[df_raw_csv['ObservationDate'].dt.month == 2]

# рассчитывает выздоровевших в каждой стране
recovered = df_february.groupby('Country/Region')['Recovered'].sum()
```

### Ответ

```python
pd.DataFrame({'Страна': recovered.keys().array, 'Выздоровевшие': recovered.values})
```

```python
for x, y in recovered.items():
    print(f'{x}\t-\t{y}')
```

## JSON:

### Задание 

Показать  уровень  фертильности  стран,  названия которых начинаются на “I”

```python
df_raw_json = pd.read_json("./Вариант 4/unece.json")

# Отбирает записи о странах начинающихся с I
df_I_countries = df_raw_json[df_raw_json['Country'].str.startswith('I')]
df_I_countries
```

```python
# Формирует pandas таблицу с нужными данными для ответа 
df_fertility = pd.DataFrame(
    {
        'Страна':       df_I_countries['Country'], 
        'Год'   :       df_I_countries['Year'],
        'Фертильность': df_I_countries['Total fertility rate']
    }
    )
```

### Ответ

```python
df_fertility
```

```python
for (x, y, z) in df_fertility.values:
    print(f'{x}\t-\t{y}\t-\t{z}')
```
